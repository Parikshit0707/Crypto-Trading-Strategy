```python
prices = []
for i in range(0, len(data['prices'])):
  prices.append(data['prices'][i][1])

print(prices)
df = pd.DataFrame()
df['prices'] = prices
df.head()
# df.set_index('date', inplace=True)
# l1 = []
# for i in range(0, len(df)):
#   l1.append(df['quote'].iloc[i]['USD']['price'])
```

```python
df['200ma'] = df['prices'].rolling(window=200).mean()
df['50ma'] = df['prices'].rolling(window=50).mean()
```

```python
print(df)
```

```python


# Calculate moving averages

# Generate signals based on moving averages
df['signal'] = 0
for i in range(0, len(df)):
  if df['prices'].iloc[i] > df.iloc[i]['200ma']:
    df['signal'].iloc[i] = 1

  elif df['prices'].iloc[i] < df.iloc[i]['50ma']:
    df['signal'].iloc[i] = -1
# df.loc[df['quote']['USD']['price'] > df['30ma'], 'signal'] = 1
# df.loc[df['quote']['USD']['price'] < df['50ma'], 'signal'] = -1

# Execute trades based on signals

amount = 0.01
position = 0
balance = 10000
trades = []

for i in range(len(df)):
    if df['signal'][i] > position:
        # Buy
        price = df['prices'].iloc[i]
        trade_cost = amount * price
        if trade_cost <= balance:
            trades.append(('buy', price, amount))
            position = 1
            balance -= trade_cost
    elif df['signal'][i] < position:
        # Sell
        price = df['prices'].iloc[i]
        trade_cost = amount * price
        if position == 1:
            trades.append(('sell', price, amount))
            position = -1
            balance += trade_cost

# Calculate profit/loss
pnl = 0
for trade in trades:
    if trade[0] == 'buy':
        pnl -= trade[1] * trade[2]
    elif trade[0] == 'sell':
        pnl += trade[1] * trade[2]
print(f'Profit/Loss: ${pnl:.2f}')

df['returns'] = np.log(df['prices'] / df['prices'].shift(1))
df['strategy_returns'] = df['signal'] * df['returns']
df['cumulative_returns'] = df['strategy_returns'].cumsum()
```

```python
average_return = df['strategy_returns'].mean()
std_return = df['strategy_returns'].std()
sharpe_ratio = average_return / std_return
print('Average return:', average_return)
print('Standard deviation of returns:', std_return)
print('Sharpe ratio:', sharpe_ratio)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load historical BTC prices

# Calculate the EMAs
df['EMA20'] = df['prices'].ewm(span=20, adjust=False).mean()
df['EMA50'] = df['prices'].ewm(span=50, adjust=False).mean()

# Generate trading signals
df['Signal'] = np.where(df['EMA20'] > df['EMA50'], 1, 0)
df['Signal'] = np.where(df['EMA20'] < df['EMA50'], -1, df['Signal'])

# Backtest the strategy
df['Return'] = df['prices'].pct_change() * df['Signal'].shift(1)
df['Cumulative Return'] = (1 + df['Return']).cumprod()

# Plot the results
plt.plot(df['Cumulative Return'])
plt.title('Exponential Moving Average Strategy for BTC')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.show()

# Print the backtesting results
total_trades = df['Signal'].count()
winning_trades = df.loc[df['Return'] > 0, 'Return'].count()
losing_trades = df.loc[df['Return'] < 0, 'Return'].count()
total_return = (df['Cumulative Return'].iloc[-1] - 1) * 100
annualized_return = (df['Cumulative Return'].iloc[-1] ** (365 / total_trades) - 1) * 100
max_drawdown = ((df['Cumulative Return'].cummax() - df['Cumulative Return']) / df['Cumulative Return'].cummax()).max() * 100
sharpe_ratio = df['Return'].mean() / df['Return'].std() * np.sqrt(365)
df['Profit/Loss'] = df['Return'] * 10000  # assuming starting capital of $10,000
total_profit_loss = df['Profit/Loss'].sum()
print('Total profit/loss: ${:.2f}'.format(total_profit_loss))
print('Total trades:', total_trades)
print('Winning trades:', winning_trades)
print('Losing trades:', losing_trades)

print('Total return: {:.2f}%'.format(total_return))
print('Annualized return: {:.2f}%'.format(annualized_return))
print('Maximum drawdown: {:.2f}%'.format(max_drawdown))
print('Sharpe ratio: {:.2f}'.format(sharpe_ratio))
```
